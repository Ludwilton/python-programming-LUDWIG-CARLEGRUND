```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
f = np.sin(x)
f
```

```python
plt.figure(dpi=100)
plt.plot(x, f)
plt.title("Sin(x)")
plt.xlabel("X")
plt.ylabel("F")
plt.show()
```

```python
# imperativt, objektorienterat
fig, ax = plt.figure(dpi=100), plt.axes()
ax.plot(x, f)
ax.set(xlim=(-6, 6), ylim=(-1.5, 1.5), xlabel="x", ylabel="f", title="sin(x)")
plt.show()
```

```python
# subplots

plt.figure(figsize=(16,5), dpi=100)
plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.title("plots")
plt.legend(("Sin(x)", "cos(x)"))

plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
plt.show()
```
