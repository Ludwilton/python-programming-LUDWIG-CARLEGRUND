# PQ formel
$ x^2 + px + q = 0$

$x = -\frac{p}{2} \pm \sqrt{(\frac{p}{2})^2 - q}$

```python
import numpy as np

q_term = [1, 2, 1]
matris_exempel = [[1, 1, 1],
                  [2, 2, 2],
                  [3, 3, 3]]

def quadratic_solver(coeff): # abc / kvadratisk formel
    a, b, c = coeff
    x1 = np.divide( - b + np.sqrt(np.square(b) - 4*a*c) , 2*a)
    x0 = np.divide( - b - np.sqrt(np.square(b) - 4*a*c) , 2*a)
    return x0, x1



print(quadratic_solver(q_term))
```

```python
# pasta konsumption
# year 0 = 1960

for year in range(1000):
    pasta_consumption = .791 * np.exp(.0526*year)
    if pasta_consumption > 15:
        print(f"Year {year+1960} is the first year that pasta consumption reaches 15kg")
        break

```

```python
import matplotlib.pyplot as plt

pasta_per_year = []

for year in range(60):
    pasta_consumption = .791 * np.exp(.0526*year)
    pasta_per_year.append(pasta_consumption)

plt.plot(range(1960,2020),pasta_per_year)
plt.xlabel("year")
plt.ylabel("pasta consumption in kg")
plt.title("pasta consumption in kg per person per year from 1960")
plt.grid(True)
```
