# Rapport Labb 3, Linjär klassifiering


Detta är en kort analysering av skillnader i klassifikation beroende på vilken lutning och skärningspunkt en linjär separationslinje har.

här följer en visualisering av linjerna enligt i) och en utskrift av antal per klass för respektive linje enligt ii)

```python
import matplotlib.pyplot as plt
import labb3 as l3


axes = plt.axes()

l3.plot_classified_points(axes=axes, color="magenta", label="k = -0.5, m = 0")

l3.plot_classified_points(k=800, m = -120, axes=axes, color="blue", label="k = 800, m = -120")

l3.plot_classified_points(k=-2, m = .16, axes=axes, color="red",label="k = -2, m=0.16" )

l3.plot_classified_points(k=-.489, axes=axes, color="green", label="k = -0.489, m = -120")
plt.grid()
plt.legend(loc="upper right")
plt.show()

```

som vi kan se i både utskriften och grafen så är fördelningen av klasserna väldigt jämna oavsett vilken typ av linje vi använder

sålänge en linje inte dras genom ett av de 2 "clusters" som finns så finns inga argument för vilken linje som går att föredra.

antalet liknande fördelningar går att hitta från en nästan helt vågrät till helt lodrät lutning sålänge skärningspunkten går genom origo.
